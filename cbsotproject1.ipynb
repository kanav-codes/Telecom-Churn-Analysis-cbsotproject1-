Approach2 with combination of trees/n_estimators and depth

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score
```

```python
n_estimators_list=[100,200,300,400,500]
max_depth_list=[5,10,15,20]
results=[]
for n_trees in n_estimators_list:
  for depth in max_depth_list:
    rf=RandomForestClassifier(n_estimators=n_trees,max_depth=depth,random_state=42,class_weight='balanced')
    rf.fit(x_train,y_train)
    y_pred=rf.predict(x_test)
    accuracy=accuracy_score(y_test,y_pred)
    recall=recall_score(y_test,y_pred)
    precision=precision_score(y_test,y_pred)
    f1=f1_score(y_test,y_pred)
    results.append({'Trees':n_trees,'Depth':depth,'Accuracy':accuracy,'Recall':recall,'Precision':precision,'F1 Score':f1})
result_df=pd.DataFrame(results)
result_df=result_df.sort_values(by=['Recall','Accuracy'],ascending=False)
print(result_df.head(20))
```

```python
segmentation_data=pd.DataFrame({'Tenure Months':x_test['Tenure Months'],'Monthly Charges':x_test['Monthly Charges'],'Total Charges':x_test['Total Charges'],'Churn Pprobability':churn_prob})
```

***Implemetation of kmeans***

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
```

```python
scaled_data=scaler.fit_transform(segmentation_data)
print(scaled_data[:5])
```

```python
from sklearn.cluster import KMeans
wcss=[]
for k in range(1,16):
  kmeans=KMeans(n_clusters=k,random_state=42,n_init='auto')
  kmeans.fit(scaled_data)
  wcss.append(kmeans.inertia_)
plt.figure(figsize=(8,6))
plt.plot(range(1,16),wcss,marker='o')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('Elbow Method For Optimal K')
plt.show()
```

```python
kmeans=KMeans(n_clusters=3,random_state=42,n_init='auto')
```

```python
clusters=kmeans.fit_predict(scaled_data)
```

```python
segmentation_data
```

```python
segmentation_data['Cluster']=clusters
```

```python
segmentation_data
```

```python
cluster_summary=segmentation_data.groupby('Cluster').mean()
```

```python
cluster_summary
```

```python

segmentation_data['Cluster'] = clusters
cluster_mapping = {
    0: 2,
    1: 0,
    2: 1
}
segmentation_data['Cluster'] = segmentation_data['Cluster'].map(cluster_mapping)


cluster_summary = segmentation_data.groupby('Cluster').mean()

print('Updated segmentation_data with remapped clusters:')
display(segmentation_data.head())

print('\nUpdated Cluster Summary with remapped clusters:')
display(cluster_summary)
```

```python
cluster_names={
    0:'Loyal Premium Customers',
    1:'Budget Loyal Customers',
    2:'High Risk Customers'
}

segmentation_data['Cluster Segment'] = segmentation_data['Cluster'].astype(int).map(cluster_names)

print("Updated segmentation_data after assigning 'Cluster Segment':")
display(segmentation_data.head())
print(f"Data type of 'Cluster' column: {segmentation_data['Cluster'].dtype}")
print(f"Data type of 'Cluster Segment' column: {segmentation_data['Cluster Segment'].dtype}")
```
